```python
import pandas as pd
import re
import numpy as np
# define data member dan booking yang ingin dijadikan acuan

%load_ext autoreload
%autoreload 2

path = r'C:\Users\fajar\Documents\Python\Data\Request Data\Salford Hall GVB_Leads_2022-08-31_2022-10-13.csv'
df = pd.read_csv(path, dtype='string')
```

```python
df.rename(columns = {'where_did_you_go_on_your_last_2_overseas_vacations_':'Destination', 
                    'would_you_rather_visit_goa,_vietnam_or_bali_':'Prefer Destination',
                    'created_time':'Date',
                    }, inplace = True)
df.columns
```

```python
df.replace('\"','', inplace=True, regex=True)
df
```

```python
datetime = df['Date'].str.split('T', expand=True)
con_date = pd.to_datetime(datetime[0])
date = con_date.dt.strftime('%d/%m/%Y')

df['Date'] = date

df
```

```python
df['Lead Source Description'] = df['Lead Source Description']+str('-')+df['campaign_name']
```

```python
df.columns
```

```python
# proper alphabetic format

df['what_are_your_occupations_'] = df['what_are_your_occupations_'].str.title()
df['Prefer Destination'] = df['Prefer Destination'].str.title()
df['full_name'] = df['full_name'].str.title()
```

```python
# clean phone number format
df['phone_number'].replace('p:\+','', regex=True, inplace=True)
df['what_ages_are_you_and_your_partner_'].replace('years|yrs','', regex=True, inplace=True)
df
```

```python
fullname = df['full_name'].str.split(' ', n=1, expand=True)
df['Firstname'] = fullname[0]
df['Lastname'] = fullname[1]
df.loc[df['Lastname'].isna(), 'Lastname'] = df['Firstname'] 
df.loc[df['Lastname'] == df['Firstname'], 'Firstname'] = ''
df.loc[df['Firstname'] == '']
```
