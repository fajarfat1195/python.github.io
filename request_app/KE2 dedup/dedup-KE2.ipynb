```python
path = r'C:\Users\fajar\Documents\Python\Data\Request Data\KE2\KE2-Dedup Master.xlsx'
df_email = pd.read_excel(path, dtype='string', sheet_name='Email')

path = r'C:\Users\fajar\Documents\Python\Data\Request Data\KE2\KE2-Dedup Master.xlsx'
df_phone = pd.read_excel(path, dtype='string', sheet_name='Phone')

path = r'C:\Users\fajar\Documents\Python\Data\Data TM for India Team - Qualified, No Members No Booking - 15 Sep 2022.xlsx'
df_tm_india_q_indonesia = pd.read_excel(path, dtype='string', sheet_name='Q-Indonesia')

path = r'C:\Users\fajar\Documents\Python\Data\Data TM for India Team - Qualified, No Members No Booking - 15 Sep 2022.xlsx'
df_tm_india_q_western = pd.read_excel(path, dtype='string', sheet_name='Q-Western')

df_tm_india_q_indonesia.fillna('empty', inplace=True)
df_tm_india_q_western.fillna('empty', inplace=True)
```

```python
# membersihkan format phone number dan email

m.clean_number(df_tm_india_q_indonesia, 'Phone')
df_tm_india_q_indonesia['Email'] = m.lowercase(df_tm_india_q_indonesia, 'Email')
m.clean_space(df_tm_india_q_indonesia, 'Email')

m.clean_number(df_tm_india_q_western, 'Phone')
df_tm_india_q_western['Email'] = m.lowercase(df_tm_india_q_western, 'Email')
m.clean_space(df_tm_india_q_western, 'Email')

m.clean_number(df_phone, 'Phone')

df_email['Email'] = m.lowercase(df_email, 'Email')
m.clean_space(df_email, 'Email')
```

```python
find_kce2_q_indonesia = df_tm_india_q_indonesia.copy()

find_kce2_q_indonesia['Same Value Find'] = 'Dedup Not Found'

find_kce2_q_indonesia.loc[find_kce2_q_indonesia['Phone'].isin(df_phone['Phone']), 'Same Value Find'] = 'Member-Mobile' 
find_kce2_q_indonesia.loc[find_kce2_q_indonesia['Email'].isin(df_email['Email']), 'Same Value Find'] = 'Member-Email'

find_kce2_q_indonesia.loc[find_kce2_q_indonesia['Same Value Find'].str.contains('Booking', flags=re.I, regex=True)]

find_kce2_q_indonesia.to_excel(r'C:\Users\fajar\Documents\Python\Data\tm_india_q_indonesia.xlsx', index=False)
```
