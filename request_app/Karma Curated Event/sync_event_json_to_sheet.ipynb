```python
import requests
import pandas as pd

import gspread
import pandas as pd
from fractions import Fraction
import re

# Pastikan permision access di menu "Share" di rubah ke anyone with the link terlebih dahulu, agar bisa mengunakan cara OAuth
# Jika Ingin lebih dibatasi menggunakan cara google service account lebih disarankan

gc = gspread.oauth(
    credentials_filename= r'..\..\Data\Google Credentials\karma-oauth.json'
)

SHEET_ID = '1k-70BJNRm01RE2S7gBXuOfqxofk1gYz4g6lPbLUCklw'
SHEET_NAME = 'Event List'
# SHEET_NAME = 'DB Test'


# Step 1: Fetch the JSON data from the URL
url = "https://karmagroup.com/karma-curated-list-json/"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse the JSON data

    # Step 2: Convert JSON data into a DataFrame
    df_json = pd.DataFrame(data)

    # Step 3: Display the DataFrame
    print("DataFrame created successfully:")

    print("DataFrame saved to 'karma_events.csv'")

else:
    print(f"Failed to fetch data. Status code: {response.status_code}")

df_json
```

```python
# get data gsheet

spreadsheet = gc.open_by_key(SHEET_ID)
worksheet = spreadsheet.worksheet(SHEET_NAME)
rows = worksheet.get_all_records()

# print(rows[:5])

# print('==============================')
# df = pd.DataFrame(rows, dtype="string")
df_sheets = pd.DataFrame(rows)
# df['Mobile Number'] = df['Mobile Number'].apply(Fraction)
df_sheets
```

```python
# Perform a left join between df_sheets and df on the 'event' column
df_join = pd.merge(df_sheets, df_json, how='left', on='event')

# Select specific columns and filter rows where 'name_y' is null (event doesn't exist in df_json, but exist in df_sheets)
df_not_exists = df_join.loc[
    df_join['name_y'].isnull(),  # Filter condition
    ['name_x', 'dates_x', 'event', 'destination_x', 'event_email_banner_x', 'event_email_banner_id_x']  # Selected columns
].copy()

# Rename columns by removing the '_x' suffix
df_not_exists = df_not_exists.rename(columns={
    'name_x': 'name',
    'dates_x': 'dates',
    'destination_x': 'destination',
    'event_email_banner_x': 'event_email_banner',
    'event_email_banner_id_x': 'event_email_banner_id'
})

# Display the resulting DataFrame
print(df_not_exists)

# Optionally, convert the filtered DataFrame to a list of lists
not_exists_list = df_not_exists.values.tolist()
print(not_exists_list)
```

```python
df_new = pd.concat([df_json, df_not_exists], ignore_index=True)
df_new
```

```python
# Update keseluruhan data, delete data yang lama terlebih dahulu
# clear all row before update
worksheet.clear()
worksheet.update([df_new.columns.values.tolist()] + df_new.values.tolist())
```
