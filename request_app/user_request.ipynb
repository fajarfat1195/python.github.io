```python
import pandas as pd
import re
import numpy as np
# define data member dan booking yang ingin dijadikan acuan

path = r'C:\Users\fajar\Documents\Python\Data\Request Data\KG - 29 Years Anniversary Giveaway 2_Leads.xlsx'
df = pd.read_excel(path, dtype='string', sheet_name='Sheet1')
df
```

```python
df.columns
```

```python
datetime = df['Created Date'].str.split('T', expand=True)
con_date = pd.to_datetime(datetime[0])
date = con_date.dt.strftime('%d/%m/%Y')

df['Created Date'] = date

df
```

```python
df['Lead Brand'] = 'Karma Group'
df['Lead Sub-Brand'] = 'Other'
df['Lead Source'] = 'Facebook Ads'
df['Lead Locations'] = 'General'

df
```

```python
# clean phone number format
df['phone_number'].replace('p:','', regex=True, inplace=True)
df
```

```python
df.loc[df['where_did_you_hear_about_this_competition?'] == 'karma_group_india', 'Lead Regions'] = 'India'
df.loc[df['where_did_you_hear_about_this_competition?'] == 'karma_group_global', 'Lead Regions'] = 'SEAP'
df.loc[df['where_did_you_hear_about_this_competition?'] == 'karma_minoan', 'Lead Regions'] = 'UK/Europe'
df.loc[df['where_did_you_hear_about_this_competition?'] == 'karma_lake_of_meneith_', 'Lead Regions'] = 'UK/Europe'
df.loc[df['where_did_you_hear_about_this_competition?'] == 'karma_salford_hall', 'Lead Regions'] = 'UK/Europe'
df.loc[df['where_did_you_hear_about_this_competition?'] == 'karma_st._martins', 'Lead Regions'] = 'UK/Europe'
df
```

```python
df['full_name'] = df['full_name'].str.title()

fullname = df['full_name'].str.split(' ', n=1, expand=True)
df['First Name'] = fullname[0]
df['Last Name'] = fullname[1]
df.loc[df['Last Name'].isna(), 'Last Name'] = df['First Name'] 
df.loc[df['Last Name'] == df['First Name'], 'First Name'] = ''
df.loc[df['First Name'] == '']
```

```python
df['marital_status'] = df['marital_status'].str.title()
```

```python
df.fillna('', inplace=True)
df.to_excel(r'C:\Users\fajar\Documents\Python\Data\KG - 29 Years Anniversary Giveaway 2_Leads-formatted.xlsx', index=False)
```
