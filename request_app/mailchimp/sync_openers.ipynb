```python
count = 0
frames = []

# mengeluarkan isi dataframe menggunakan perulangan
for index, row in df_campaign.iterrows():
        print(f"Retrieveing openers data {row['id']}-{row['settings']}")

        # Mencari jumlah page openers dari campaign name
        globals()['df_campaign_openers%s' % count] = mm.get_data_list(api_key, server, row['id'])

        frames.append(globals()['df_campaign_openers%s' % count])
        count += 1
if len(frames) > 0:
        df_campaign_openers = pd.concat(frames, ignore_index=True)
        mask = df_campaign_openers.loc[df_campaign_openers['email_address'].str.contains('karmagroup|karmaclub|karmaexperience', regex=True)]
        df_campaign_openers.drop(mask.index, inplace=True)
else:
        df_campaign_openers = pd.DataFrame({'Campaign Name': [''], 'Openers': ['']})   
```
