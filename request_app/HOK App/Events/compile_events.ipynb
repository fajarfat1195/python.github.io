```python
# import OS module
import os
import pandas as pd
 
# Get the list of all files and directories
# dir_path = "C://Users//Vanshi//Desktop//gfg"
dir_path = r"C:\Users\fajar\Documents\Python\Data\Request Data\Hok Event Tmp"

# dir_list = os.listdir(dir_path)
 
# print("Files and directories in '", dir_path, "' :")
 
# # prints all files
# print(dir_list)

frames = []

# itterate file di dalam directory
for path in os.listdir(dir_path):

    # cek apakah di dalam directory ada file
    if os.path.isfile(os.path.join(dir_path, path)):
        if ".xlsx" in path:
            print(path)
            df = pd.read_excel(dir_path+'\\'+path, dtype='string')
            df['filename'] = path
            frames.append(df)

        if ".csv" in path:
            print(path)
            df = pd.read_csv(dir_path+'\\'+path, dtype='string', on_bad_lines='skip')
            df['filename'] = path
            frames.append(df)
    else:
        print('file doesn\'t exist')

result = pd.concat(frames, ignore_index=True)
result
```

```python
result.to_excel(r'C:\Users\fajar\Documents\Python\Data\Request Data\Hok Event Tmp\compiled_file.xlsx', index=False)
```

```python
result.rename(columns={'02- Mobile No.':'02- Mobile',
                       '03- Mobile No.':'03- Mobile'
                       }, inplace=True)
result.columns
```

```python
df_1 = result.reindex(columns=['Timestamp', 'Source', '01- First Name', '01- Last Name', '01- Email',
       '01- Mobile', '01- Dietary Requirements','Member Details', 'Submitted From URL', 'filename']).copy()

df_2 = result.reindex(columns=['Timestamp', 'Source', '02- First Name', '02- Last Name', '02- Email',
       '02- Mobile', '02- Dietary Requirements','Member Details', 'Submitted From URL', 'filename']).copy()

df_3 = result.reindex(columns=['Timestamp', 'Source', '03- First Name', '03- Last Name', '03- Email',
       '03- Mobile', '03- Dietary Requirements','Member Details','Submitted From URL', 'filename']).copy()

df_4 = result.reindex(columns=['Timestamp', 'Source', '04- First Name', '04- Last Name', '04- Email',
       '04- Mobile', '04- Dietary Requirements','Member Details','Submitted From URL', 'filename']).copy()

df_5 = result.reindex(columns=['Timestamp', 'Source', '05- First Name', '05- Last Name', '05- Email',
       '05- Mobile', '05- Dietary Requirements','Member Details','Submitted From URL', 'filename']).copy()
```

```python
df_2.loc[df_2['02- Email'].notna() | df_2['02- Mobile'].notna(),'Source'] = 'Referral'
df_2.drop(df_2.loc[df_2['02- Email'].isna() & df_2['02- Mobile'].isna()].index, inplace=True)
df_2.reset_index(drop=True, inplace=True)

df_3.loc[df_3['03- Email'].notna() | df_3['03- Mobile'].notna(),'Source'] = 'Referral'
df_3.drop(df_3.loc[df_3['03- Email'].isna() & df_3['03- Mobile'].isna()].index, inplace=True)
df_3.reset_index(drop=True, inplace=True)

df_4.loc[df_4['04- Email'].notna() | df_4['04- Mobile'].notna(),'Source'] = 'Referral'
df_4.drop(df_4.loc[df_4['04- Email'].isna() & df_4['04- Mobile'].isna()].index, inplace=True)
df_4.reset_index(drop=True, inplace=True)

df_5.loc[df_5['05- Email'].notna() | df_5['05- Mobile'].notna(),'Source'] = 'Referral'
df_5.drop(df_5.loc[df_5['05- Email'].isna() & df_5['05- Mobile'].isna()].index, inplace=True)
df_5.reset_index(drop=True, inplace=True)
```

```python
df_1.rename(columns={'01- First Name':'First Name', 
                     '01- Last Name':'Last Name',
                     '01- Email':'Email',
                     '01- Mobile':'Mobile', 
                     '01- Dietary Requirements':'Dietary Requirements',
                       }, inplace=True)

df_2.rename(columns={'02- First Name':'First Name', 
                     '02- Last Name':'Last Name',
                     '02- Email':'Email',
                     '02- Mobile':'Mobile', 
                     '02- Dietary Requirements':'Dietary Requirements'
                       }, inplace=True)

df_3.rename(columns={'03- First Name':'First Name', 
                     '03- Last Name':'Last Name',
                     '03- Email':'Email',
                     '03- Mobile':'Mobile', 
                     '03- Dietary Requirements':'Dietary Requirements'
                       }, inplace=True)

df_4.rename(columns={'04- First Name':'First Name', 
                     '04- Last Name':'Last Name',
                     '04- Email':'Email',
                     '04- Mobile':'Mobile', 
                     '04- Dietary Requirements':'Dietary Requirements'
                       }, inplace=True)

df_5.rename(columns={'05- First Name':'First Name', 
                     '05- Last Name':'Last Name',
                     '05- Email':'Email',
                     '05- Mobile':'Mobile', 
                     '05- Dietary Requirements':'Dietary Requirements'
                       }, inplace=True)
```

```python
df_final = pd.concat([df_1,df_2,df_3,df_4,df_5])
df_final.drop_duplicates(subset=['Email','filename'], inplace=True)
df_final.reset_index(drop=True, inplace=True)
df_final['First Name'] = df_final['First Name'].str.title()
df_final['Last Name'] = df_final['Last Name'].str.title()
df_final['Email'] = df_final['Email'].str.lower()
```

```python
df_final['Member Details'].replace(' ','', regex=True, inplace=True)
df_final['Member Details'].replace('[^0-9]','', regex=True, inplace=True) 
```

```python
df_final.rename(columns={'Source':'Guest Type',
                         'Member Details':'Odyssey_Member_ID'
                      }, inplace=True)
```

```python
df_final.to_csv(r'C:\Users\fajar\Documents\Python\Data\Request Data\Hok Event Tmp\compiled_normalize.csv', index=False)
```
