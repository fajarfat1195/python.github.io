```python
# import OS module
import os
import pandas as pd
 
# Get the list of all files and directories
# dir_path = "C://Users//Vanshi//Desktop//gfg"
dir_path = r"G:\.shortcut-targets-by-id\1BKe3lpZLP3aj_PPa_Bai9V8nzzywN1RZ\_Toni\Qquence"

# dir_list = os.listdir(dir_path)
 
# print("Files and directories in '", dir_path, "' :")
 
# # prints all files
# print(dir_list)

frames = []

# itterate file di dalam directory
for path in os.listdir(dir_path):

    # cek apakah di dalam directory ada file
    if os.path.isfile(os.path.join(dir_path, path)):
        if ".xlsx" in path:
            print(path)
            df = pd.read_excel(dir_path+'\\'+path, dtype='string')
            df['filename'] = path
            frames.append(df)

        if ".csv" in path:
            print(path)
            df = pd.read_csv(dir_path+'\\'+path, dtype='string', on_bad_lines='skip')
            df['filename'] = path
            frames.append(df)
    else:
        print('file doesn\'t exist')

result = pd.concat(frames, ignore_index=True)
result
```

```python
result.loc[result['EMAIL'].isna(),'EMAIL'] = result['EMAIL ADDRESS']
result.loc[result['PHONE NUMBER'].isna(),'PHONE NUMBER'] = result['PHONE NUMBER.']
result.loc[result['PHONE NUMBER'].isna(),'PHONE NUMBER'] = result['PHONE']
result.loc[result['FULL NAME'].isna(),'FULL NAME'] = result['FIRST NAME']+' '+result['LAST NAME']

result.loc[result['OPT-IN'].isna(),'OPT-IN'] = result['ORT-IN']

name = result['FULL NAME'].str.split(" ", n=1, expand=True)

result['FIRST NAME'] = name[0]
result['LAST NAME'] = name[1]

result['PHONE NUMBER'].replace(' ','', regex=True, inplace=True)
result['PHONE NUMBER'].replace('[^0-9]','', regex=True, inplace=True) 

result.loc[result['PHONE NUMBER'].str.len() < 5, 'PHONE NUMBER'] = ''
result.loc[~result['EMAIL'].str.contains('@'), 'EMAIL'] = ''
```
