```python
import pandas as pd
import re

import past_guest_module as pm

import sys
sys.path.append('../zoho_data_app/')

import main_module as m

%load_ext autoreload
%autoreload 2
```

```python
path = r'C:\Users\fajar.fatoni\Documents\Python\Data\Past Guest Data\View Point\2025\Sep\31 Aug - 08 Sep\CheckOutReport 31 Aug - 08 Sep.xlsx'
# path = r'C:\Users\fajar.fatoni\Documents\Python\Data\CheckOutReport KBorgo.xlsx'
df_viewpoint = pd.read_excel(path, dtype='string', sheet_name='Sheet1')
count_data_awal = df_viewpoint.shape[0]
df_viewpoint
```

```python
df_viewpoint.columns
```

```python
df_viewpoint.fillna('empty', inplace=True)

df_viewpoint['Contact Type'] = 'empty'
df_viewpoint['Odyssey Members'] = 'NO'

df_viewpoint['Lead Brand'] = 'Karma Resorts'
df_viewpoint['Lead Sub-Brand'] = 'Other'
df_viewpoint['Lead Source'] = 'Past Guests'
df_viewpoint['Lead Locations'] = df_viewpoint['resortname']
df_viewpoint['Lead Source Description'] = 'Viewpoint'
```

```python
df_viewpoint.loc[df_viewpoint['Opt In'] == 'YES', 'Exclude From Mailing'] = 'NO'
df_viewpoint.loc[df_viewpoint['Opt In'] == 'NO', 'Exclude From Mailing'] = 'YES'
```

```python
df_viewpoint.loc[df_viewpoint['lastname'] == 'empty', 'lastname'] = '-'
```

```python
df_viewpoint['Title'].replace('empty', '', inplace=True)
df_viewpoint['Firstname'].replace('empty', '', inplace=True)
df_viewpoint['lastname'].replace('empty', '', inplace=True)

df_viewpoint['Title'] = df_viewpoint['Title'].str.title()
df_viewpoint['Firstname'] = df_viewpoint['Firstname'].str.title()
df_viewpoint['lastname'] = df_viewpoint['lastname'].str.title()

df_viewpoint
```

```python
m.clean_number(df_viewpoint, 'Phone')
m.clean_number(df_viewpoint, 'Mobile')
m.replace_not_valid_to_empty(df_viewpoint, 'Phone')
m.replace_not_valid_to_empty(df_viewpoint, 'Mobile')

df_viewpoint.loc[df_viewpoint['Phone'].str.contains('00000'), 'Phone'] = 'empty'
df_viewpoint.loc[df_viewpoint['Mobile'].str.contains('00000'), 'Mobile'] = 'empty'

m.clean_space(df_viewpoint, 'Email')
df_viewpoint['Email'] = m.lowercase(df_viewpoint, 'Email')
m.email_correction(df_viewpoint, 'Email')
```

```python
df_viewpoint
```

```python
df_viewpoint.loc[(df_viewpoint['Email'] != 'empty') & ((df_viewpoint['Phone'] != 'empty') | (df_viewpoint['Mobile'] != 'empty')), 'Contact Type'] = 'Email and Phone'
df_viewpoint.loc[(df_viewpoint['Email'] != 'empty') & ((df_viewpoint['Phone'] == 'empty') & (df_viewpoint['Mobile'] == 'empty')), 'Contact Type'] = 'Email Only'
df_viewpoint.loc[(df_viewpoint['Email'] == 'empty') & ((df_viewpoint['Phone'] != 'empty') | (df_viewpoint['Mobile'] != 'empty')), 'Contact Type'] = 'Phone Only'

df_viewpoint.loc[df_viewpoint['Contact Type'] == 'empty'].to_csv(r'C:\Users\fajar.fatoni\Documents\Python\Data\contact_type_empty_viewpoint.csv', index=False)

count_contact_type_empty = df_viewpoint.loc[df_viewpoint['Contact Type'] == 'empty'].shape[0]

df_viewpoint.drop(df_viewpoint.loc[df_viewpoint['Contact Type'] == 'empty'].index, inplace=True)
df_viewpoint.reset_index(drop=True, inplace=True)

count_after_contact_type_empty = df_viewpoint.shape[0]

df_viewpoint
```

```python
# Cek Email Karma
email_karma = m.find_karma(df_viewpoint, 'Email')
count_email_karma = email_karma.shape[0]
email_karma.to_csv(r'C:\Users\fajar.fatoni\Documents\Python\Data\email_deleted_viewpoint.csv', index=False)

email_karma
```

```python
# delete email Karma
count_after_clean_karma = m.delete_karma(df_viewpoint, 'Email')

df_viewpoint
```

```python
# Cek leads test
test_leads = df_viewpoint.loc[
    
        df_viewpoint['Email'].str.contains('^test@', flags=re.I, regex=True) |
        df_viewpoint['Email'].str.contains('\+test', flags=re.I, regex=True) |
    
        df_viewpoint['Firstname'].str.contains('^test$', flags=re.I, regex=True) |
        df_viewpoint['lastname'].str.contains('^test$', flags=re.I, regex=True) |

        df_viewpoint['Firstname'].str.contains('test 1', flags=re.I, regex=True) |
        df_viewpoint['lastname'].str.contains('test 2', flags=re.I, regex=True)|
        df_viewpoint['Firstname'].str.contains('test1', flags=re.I, regex=True) |
        df_viewpoint['lastname'].str.contains('test2', flags=re.I, regex=True)|

        df_viewpoint['Firstname'].str.contains(' test ', flags=re.I, regex=True) | 
        df_viewpoint['lastname'].str.contains(' test ', flags=re.I, regex=True) | 
        df_viewpoint['lastname'].str.contains(' test$', flags=re.I, regex=True)
    ]

count_leads_test = test_leads.shape[0]

test_leads.to_csv(r'C:\Users\fajar.fatoni\Documents\Python\Data\test_leads_viewpoint.csv', index=False)
test_leads
```

```python
df_viewpoint.drop(test_leads.index, inplace=True)
df_viewpoint.reset_index(drop=True, inplace=True)

count_after_dedup_test = df_viewpoint.shape[0]
count_after_dedup_test
```
