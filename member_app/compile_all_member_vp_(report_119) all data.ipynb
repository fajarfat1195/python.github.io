```python
import sys

sys.path.append('../zoho_data_app')

import main_module as m
import zoho_filter_module as fil
import pandas as pd
import re
import numpy as np

import glob

%load_ext autoreload
%autoreload 2

# download semua report 63 di vp
# download 2x berdasarkan servicing office yang non-india dan india 

file_loc = r'C:\Users\fajar.fatoni\Documents\Python\Data\1184820-MemberContactDetails.xlsx'
```

```python
excel_file = pd.ExcelFile(file_loc)
sheet_names = excel_file.sheet_names
frames = []
count = 0
for sheet in sheet_names:
    count += 1
    print(f"Compiling {sheet}")
    if sheet == 'Sheet1':
        globals()['df%s' % count]= pd.read_excel(file_loc, sheet_name=sheet, dtype='string')
        globals()['column'] = globals()['df%s' % count].columns.to_list()
    else:
        globals()['df%s' % count]= pd.read_excel(file_loc, sheet_name=sheet, dtype='string', header=None)
        globals()['df%s' % count].columns = globals()['column']

    frames.append(globals()['df%s' % count])

result = pd.concat(frames, ignore_index=True)

result.drop(result.loc[result['Account ID'] == 'krg'].index, inplace=True)
result
```

```python
df_final = pd.concat([result], ignore_index=True)
df_final.fillna('empty', inplace=True)
df_final
```

```python
# data phone, mobile dam email member yang ada di master data (wash member)

df_final['Email'] = m.lowercase(df_final, 'Email')
m.clean_space(df_final, 'Email')
m.clean_not_valid_email_format(df_final, 'Email')
```

```python
df_final.replace('empty', '', inplace=True)
df_final
```

```python
df_final.loc[df_final['Country'] == 'India'].to_csv(r'C:\Users\fajar.fatoni\Documents\Python\Data\compiled_member.csv', index=False)
```
