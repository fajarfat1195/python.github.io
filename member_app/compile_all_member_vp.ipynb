```python
import sys

sys.path.append('../zoho_data_app')

import main_module as m
import zoho_filter_module as fil
import pandas as pd
import re
import numpy as np

import glob

%load_ext autoreload
%autoreload 2

# download semua report 63 di vp
# download 2x berdasarkan servicing office yang non-india dan india 

file_loc_india = r'C:\Users\fajar\Documents\Python\Data\Member Data\vp\report 63 all details\India.xlsx'
file_loc_non_india = r'C:\Users\fajar\Documents\Python\Data\Member Data\vp\report 63 all details\Non-India.xlsx'
```

```python
excel_file = pd.ExcelFile(file_loc_india)
sheet_names = excel_file.sheet_names
frames = []
count = 0
for sheet in sheet_names:
    count += 1
    print(f"Compiling {sheet}")
    if sheet == 'Sheet1':
        globals()['df%s' % count]= pd.read_excel(file_loc_india, sheet_name=sheet, dtype='string')
        globals()['column'] = globals()['df%s' % count].columns.to_list()
    else:
        globals()['df%s' % count]= pd.read_excel(file_loc_india, sheet_name=sheet, dtype='string', header=None)
        globals()['df%s' % count].columns = globals()['column']

    frames.append(globals()['df%s' % count])

result_india = pd.concat(frames, ignore_index=True)

result_india.drop(result_india.loc[result_india['AccountID'] == 'krg'].index, inplace=True)
result_india
```

```python
excel_file = pd.ExcelFile(file_loc_non_india)
sheet_names = excel_file.sheet_names
frames = []
count = 0
for sheet in sheet_names:
    count += 1
    print(f"Compiling {sheet}")
    if sheet == 'Sheet1':
        globals()['df%s' % count]= pd.read_excel(file_loc_non_india, sheet_name=sheet, dtype='string')
        globals()['column'] = globals()['df%s' % count].columns.to_list()
    else:
        globals()['df%s' % count]= pd.read_excel(file_loc_non_india, sheet_name=sheet, dtype='string', header=None)
        globals()['df%s' % count].columns = globals()['column']

    frames.append(globals()['df%s' % count])

result = pd.concat(frames, ignore_index=True)

result.drop(result.loc[result['AccountID'] == 'krg'].index, inplace=True)
result
```

```python
df_final = pd.concat([result_india, result], ignore_index=True)
df_final.fillna('empty', inplace=True)
df_final
```

```python
# data phone, mobile dam email member yang ada di master data (wash member)

m.clean_number(df_final, 'Phone')
m.clean_number(df_final, 'Mobile')
m.replace_not_valid_to_empty(df_final, 'Phone')
m.replace_not_valid_to_empty(df_final, 'Mobile')
df_final['Email'] = m.lowercase(df_final, 'Email')
m.clean_space(df_final, 'Email')
m.clean_not_valid_email_format(df_final, 'Email')
```

```python
df_final.replace('empty', '', inplace=True)
df_final
```
