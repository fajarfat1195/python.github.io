```python
def next_available_row(worksheet):
    str_list = worksheet.col_values(1)
    return str(len(str_list)+1)

def convert_to_list(value):
  result = [value]
  return result
```

```python
# get data gsheet

spreadsheet = gc.open_by_key(SHEET_ID)
worksheet = spreadsheet.worksheet(SHEET_NAME)
rows = worksheet.get_all_records()

# print(rows[:5])

# print('==============================')
df = pd.DataFrame(rows)
df
```

```python
# df.loc[df['Campaign Name'].str.contains('fbi', flags=re.I, regex=True)]
df['Campaign Name'] = df['Campaign Name'].str.replace('fbi', 'fb', flags=re.I, regex=True)
```

```python
df['Programme'] = ''

c_salak = df['Campaign Name'].str.contains('salak', flags=re.I, regex=True)
c_kandara = df['Campaign Name'].str.contains('kandara', flags=re.I, regex=True)
c_merapi = df['Campaign Name'].str.contains('merapi', flags=re.I, regex=True)
c_beach = df['Campaign Name'].str.contains('beach', flags=re.I, regex=True)
c_discovery = df['Campaign Name'].str.contains('discovery', flags=re.I, regex=True)

c_flybuy = df['Campaign Name'].str.contains('flybuy|fly buy|fb ', flags=re.I, regex=True)
c_drivebuy = df['Campaign Name'].str.contains('drivebuy|drive buy', flags=re.I, regex=True)
c_coldline = df['Campaign Name'].str.contains('coldline', flags=re.I, regex=True)
c_roadshow = df['Campaign Name'].str.contains('roadshow', flags=re.I, regex=True)

df.loc[c_salak, 'Programme'] = 'Karma Salak'
df.loc[c_kandara, 'Programme'] = 'Karma Kandara'
df.loc[c_merapi, 'Programme'] = 'Karma Merapi'
df.loc[c_beach, 'Programme'] = 'Karma Beach'
df.loc[c_discovery, 'Programme'] = 'Karma Discovery'

c_prog_is_empty = df['Programme'].str.len() < 1
    
df.loc[(c_flybuy) & (c_prog_is_empty), 'Programme'] = 'Fly Buy'
df.loc[(c_drivebuy) & (c_prog_is_empty), 'Programme'] = 'Drive Buy'
df.loc[(c_coldline) & (c_prog_is_empty), 'Programme'] = 'Coldline'
df.loc[(c_roadshow) & (c_prog_is_empty), 'Programme'] = 'Roadshow'
df
```

```python
# Applying lambda function to multiple rows using df.apply()

df["Programme"] = df.apply(
    lambda x: convert_to_list(x['Programme']) ,
      
    axis=1,
)
df.loc[c_flybuy]
```
