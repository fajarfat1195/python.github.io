```python
# get data gsheet

spreadsheet = gc.open_by_key(SHEET_ID)
worksheet = spreadsheet.worksheet(SHEET_NAME)
rows = worksheet.get_all_records()

# print(rows[:5])

# print('==============================')
df = pd.DataFrame(rows)
df['Mobile Number'] = df['Mobile Number'].apply(Fraction)
df['Lead Location'] = SHEET_NAME
```

```python
# get data gsheet

spreadsheet = gc.open_by_key(SHEET_ID)
worksheet = spreadsheet.worksheet(SHEET_NAME_3)
rows = worksheet.get_all_records()

# print(rows[:5])

# print('==============================')
df3 = pd.DataFrame(rows)
df3['Mobile Number'] = df3['Mobile Number'].apply(Fraction)
df3['Lead Location'] = SHEET_NAME_3
```

```python
df_all = pd.concat([df,df2,df3], ignore_index=True)
df_all
```

```python
df_organic = df_all.loc[df_all['Source'].str.contains('organic', regex=True, flags=re.I)].copy()
df_organic
```

```python
df_organic['Name'] = df_organic['Name'].str.title()
split_name = df_organic['Name'].str.split(' ', expand=True, n=2)
df_organic['First Name'] =  split_name[0]
df_organic['Last Name'] =  split_name[1]

df_organic.loc[df_organic['First Name'] == 'None', 'First Name'] = df_organic['Name']
df_organic.loc[df_organic['Last Name'] == 'None', 'Last Name'] = '-'
```
