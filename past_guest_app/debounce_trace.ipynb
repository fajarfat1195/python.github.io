```python
from sqlalchemy import create_engine

from datetime import datetime
from uuid import uuid4
    
# 'mysql+pymysql://mysql_user:mysql_password@mysql_host/mysql_db'
engine_str = 'mysql+pymysql://root:@localhost/past_guest_tmp'
engine = create_engine(engine_str)

import pprint
import pandas as pd
import numpy as np
from normalization import setDateTime
from debounce import Debounce
deb = Debounce()
```

```python
def debounce_check(email):
    if email != 'empty_email':
        deb_data = deb.validonce(email)
        reason = deb_data['debounce']['reason']
        result = deb_data['debounce']['result']
        status_code = deb_data['debounce']['code']
        role = deb_data['debounce']['role']
        free = deb_data['debounce']['free_email']

        now = datetime.now()
        dt_string = now.strftime("%Y-%m-%d %H:%M:%S")


        engine.execute("INSERT INTO tb_debounce (email, result, reason, status_code, role, free, created_time) VALUES ('"+email+"', '"+result+"', '"+reason+"', '"+status_code+"', '"+role+"', '"+free+"','"+dt_string+"')")
    else:
        reason = ''
    return reason
```

```python
path = r'C:\Users\fajar\Documents\Python\Data\normalize_viewpoint.csv'
df_pg_vp = pd.read_csv(path, dtype='string', encoding='utf8')
df_pg_vp['Email'].fillna('empty_email', inplace=True)
df_pg_vp['Email_status'] = df_pg_vp.apply(
    lambda x: debounce_check(x['Email']),
    axis=1,
)
```
