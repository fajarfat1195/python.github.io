```python
def con_fraction(phone):
    if phone == '':
        new_phone = ''
    else:
        new_phone = Fraction(phone)
    return new_phone

def con_str(phone):
    if phone == '':
        new_phone = ''
    else:
        new_phone = '+'+str(int(phone))
    return new_phone

def con_number(number):
    if number == '':
        new_number = ''
    else:
        new_number = int(number)
    return new_number
```

```python
frames = [df_check_out, df_cancelled, df_viewpoint]
new_df = pd.concat(frames, ignore_index=True)

new_df.fillna('', inplace=True)

new_df['Destination'] = new_df['Lead_Locations']
new_df['Do_Not_Mail'] = new_df['Email_Opt_Out']


new_df['Phone'] = new_df.apply(
lambda x: con_str(x['Phone']),
    axis=1,
)

new_df['Mobile'] = new_df.apply(
    lambda x: con_str(x['Mobile']),
    axis=1,
)

# new_df['Phone'] = new_df.apply(
# lambda x: con_fraction(x['Phone']),
#     axis=1,
# )

# new_df['Mobile'] = new_df.apply(
#     lambda x: con_fraction(x['Mobile']),
#     axis=1,
# )

new_df['Adult'] = new_df.apply(
    lambda x: con_number(x['Adult']),
    axis=1,
)

new_df['Children'] = new_df.apply(
    lambda x: con_number(x['Children']),
    axis=1,
)
```

```python
dob_con_date = pd.to_datetime(new_df['Birthdate'])
dob_date = dob_con_date.dt.strftime('%d/%m/%Y')

arr_con_date = pd.to_datetime(new_df['Arrival_Date'])
arr_date = arr_con_date.dt.strftime('%d/%m/%Y')

dep_con_date = pd.to_datetime(new_df['Departure_Date'])
dep_date = dep_con_date.dt.strftime('%d/%m/%Y')

new_df['Birthdate'] = dob_date
new_df['Arrival_Date'] = arr_date
new_df['Departure_Date'] = dep_date
new_df
```
