```python
import sys
sys.path.append('../zoho_data_app')

import main_module as m
import pandas as pd
import re

import gspread

%load_ext autoreload
%autoreload 2
```

```python
gc = gspread.oauth(
    credentials_filename= r'C:\Users\fajar.fatoni\Documents\Python\Data\Google Credentials\karma-oauth.json'
)

SHEET_ID = '1FZ4bgWOEiVsbICA0WcG7DmBmjqA58v7ZdpiRZrARCqY'
SHEET_NAME_KLOM = 'KLOM'
SHEET_NAME_KSH = 'KSH'
SHEET_NAME_KB = 'KB'
SHEET_NAME_LSQ = 'Done in LSQ'
SHEET_NAME_API = 'API'

# get data gsheet

spreadsheet = gc.open_by_key(SHEET_ID)

worksheet_klom= spreadsheet.worksheet(SHEET_NAME_KLOM)
rows_klom = worksheet_klom.get_all_records()
df_klom = pd.DataFrame(rows_klom)
df_klom['Lead Location'] = 'Karma Lake Of Menteith'

worksheet_ksh= spreadsheet.worksheet(SHEET_NAME_KSH)
rows_ksh = worksheet_ksh.get_all_records()
df_ksh = pd.DataFrame(rows_ksh)
df_ksh['Lead Location'] = 'Karma Salford Hall'

worksheet_kb= spreadsheet.worksheet(SHEET_NAME_KB)
rows_kb = worksheet_kb.get_all_records()
df_kb = pd.DataFrame(rows_kb)
df_kb['Lead Location'] = 'Karma Bavaria'

worksheet_lsq= spreadsheet.worksheet(SHEET_NAME_LSQ)
rows_lsq = worksheet_lsq.get_all_records()
df_lsq = pd.DataFrame(rows_lsq)

worksheet_api= spreadsheet.worksheet(SHEET_NAME_API)
rows_api = worksheet_api.get_all_records()
df_api = pd.DataFrame(rows_api)

df_klom
```

```python
df_offline_join = pd.concat([df_klom,df_ksh,df_kb])
df_offline_join.loc[df_offline_join['Email'] == '','Email'] = 'empty'
df_offline_join.loc[df_offline_join['Phone'] == '','Email'] = 'empty'
count_data_gabungan = df_offline_join.shape[0]
df_offline_join
```

# ====================== Awal Filter Umum ====================== 

```python
# data phone, mobile dam email member yang ada di master data (wash member)

m.clean_number(df_offline_join, 'Phone')
m.replace_not_valid_to_empty(df_offline_join, 'Phone')
df_offline_join['Email'] = m.lowercase(df_offline_join, 'Email')
m.clean_space(df_offline_join, 'Email')

m.clean_number(df_lsq, 'Phone')
df_lsq['Email'] = m.lowercase(df_lsq, 'Email')
m.clean_space(df_lsq, 'Email')

m.clean_number(df_api, 'Phone')
df_lsq['Email'] = m.lowercase(df_api, 'Email')
m.clean_space(df_api, 'Email')
```

# ====================== Akhir Filter Umum ====================== 

```python
# hitung jumlah duplicate
count_duplicate_general = df_offline_join.loc[df_offline_join.duplicated(subset=['Email'])].shape[0]

df_offline_join.loc[df_offline_join.duplicated(subset=['Email'])]
```

```python
df_offline_join.drop_duplicates(subset=['Email'], inplace=True)
df_offline_join.reset_index(drop=True, inplace=True)

# hitung jumlah setelah value duplicate dihapus
count_after_dedup_duplicate = df_offline_join.shape[0]
df_offline_join
```

```python
# data phone, mobile dam email member yang ada di master data 
count_wash_before = df_offline_join.loc[
    df_offline_join['Phone'].isin(df_lsq['Phone']) | 
    df_offline_join['Email'].isin(df_lsq['Email'])| 

    df_offline_join['Phone'].isin(df_api['Phone']) | 

    df_offline_join['Email'].isin(df_api['Email'])

].shape[0]

df_offline_join.drop(df_offline_join.loc[
    df_offline_join['Phone'].isin(df_lsq['Phone']) | 
    df_offline_join['Email'].isin(df_lsq['Email'])| 

    df_offline_join['Phone'].isin(df_api['Phone']) | 

    df_offline_join['Email'].isin(df_api['Email'])

    ].index, inplace=True)
    
df_offline_join.reset_index(drop=True, inplace=True)

count_after_wash_before = df_offline_join.shape[0]
```

```python
# data phone, mobile dam email member yang ada di master data 
count_wash_contact_empty = df_offline_join.loc[
    (df_offline_join['Email'] == 'empty') & 
    (df_offline_join['Phone'] == 'empty')
    
].shape[0]

df_offline_join.drop(df_offline_join.loc[
    (df_offline_join['Email'] == 'empty') & 
    (df_offline_join['Phone'] == 'empty')
    ].index, inplace=True)
    
df_offline_join.reset_index(drop=True, inplace=True)

count_after_wash_contact_empty = df_offline_join.shape[0]
```

```python
print("==========================================================")
print("Summary")
print("==========================================================")
print("JumQUlah Filter Leads Awal : "+str(count_data_gabungan))
print("Jumlah Duplicate (Phone, Mobile dan Email) : "+str(count_duplicate_general))
print("Jumlah Setelah Clean Duplicate : "+str(count_after_dedup_duplicate))
print("Jumlah Wash Before : "+str(count_wash_before))
print("Jumlah Setelah Wash Member : "+str(count_after_wash_before))
print("Jumlah Wash Empty Contact : "+str(count_wash_contact_empty))
print("Jumlah Setelah Wash Empty Contact : "+str(count_after_wash_contact_empty))
```

```python
df_final = pd.DataFrame()
df_final['Created Time'] = df_offline_join['First visit date'].copy()
df_final['Lead Location'] = df_offline_join['Lead Location'].copy()
df_final[['First Name', 'Last Name']] = df_offline_join['Full name'].str.split(' ', n=1, expand=True)
df_final['Email'] = df_offline_join['Email'].copy()
df_final['Phone'] = df_offline_join['Phone'].copy()
df_final['Lead Brand'] = 'Timeshare Marketing'
df_final['Lead Sub-Brand'] = 'Other'
df_final['Lead Source'] = 'Wifi'
df_final['Lead Source Description'] = 'Beambox'
df_final['Email Opt Out'] = False
df_final['Opt In'] = True
df_final['Do Not Contact'] = False
df_final['Do Not Mail'] = False

df_final['Phone'] = df_final['Phone'].astype("string")

df_final['Phone'].replace('^440','44', regex=True, inplace=True)
df_final
```

```python
df_final.replace('empty', '', inplace=True, regex=True)
df_final.loc[df_final['Phone'].str.len() < 3, 'Phone'] = ''
```
