```python
# import OS module
import os
import pandas as pd
 
# Get the list of all files and directories
# dir_path = "C://Users//Vanshi//Desktop//gfg"
dir_path = r"C:\Users\fajar\Documents\Python\Data\TM Backup"

frames = []

# itterate file di dalam directory
for path in os.listdir(dir_path):

    # cek apakah di dalam directory ada file
    if os.path.isfile(os.path.join(dir_path, path)):
        if ".xlsx" in path:
            print(path)
            frames.append(pd.read_excel(dir_path+'\\'+path, dtype='string'))
    else:
        print('file doesn\'t exist')

result = pd.concat(frames, ignore_index=True)
result
```

```python
result['First Name'] = result['First Name'].str.title()
result['Last Name'] = result['Last Name'].str.title()
```

```python
result.groupby('Lead Sub-Brand').count()
```

```python
df_empty_birth = result.loc[result['Birthdate'].isna()].copy()
df_empty_birth.reset_index(drop=True, inplace=True)
df_empty_birth.sort_values('Birthdate', ascending=True)
```

```python
df_empty_birth.groupby('Lead Sub-Brand').count()
```

```python
df_not_empty_birth = result.loc[~result['Birthdate'].isna()].copy()
df_not_empty_birth.reset_index(drop=True, inplace=True)
df_not_empty_birth
```

```python
df_not_empty_birth.sort_values('Birthdate')
```

```python
df_not_empty_birth.groupby('Lead Sub-Brand').count()
```

```python
year_of_birth = df_not_empty_birth['Birthdate'].str.split('-', expand=True)
df_not_empty_birth['Check Year'] = year_of_birth[0].astype(int)
```

```python
df_not_empty_birth[['Email','Birthdate','Check Year']].sort_values('Birthdate')
```

```python
df_not_empty_birth.dtypes
```

```python
df_valid_birthyear = df_not_empty_birth.loc[(df_not_empty_birth['Check Year'] > 1930) & (df_not_empty_birth['Check Year'] < 2023)].copy()
df_valid_birthyear.reset_index(drop=True, inplace=True)
df_valid_birthyear
```

```python
df_valid_birthyear.groupby('Lead Sub-Brand').count()
```
