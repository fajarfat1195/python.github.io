```python
path = r'C:\Users\fajar\Documents\Python\Data\Debounce\zoho campaign\compiled\Email Debounce Campaign Master.xlsx'
df_message = pd.read_excel(path, dtype='string', sheet_name='Message Delete')
df_message['Error Message'] = df_message['Error Message'].str.strip()
df_message_tr = df_message.loc[df_message['Delete'] == 'True'].copy()
```

```python
joined_df = pd.merge(debounce_df,
                     df_message_tr,
                     left_on='Bounce reason',
                     right_on='Error Message',
                     how ='left')
joined_df
```
