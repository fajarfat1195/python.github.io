```python
# hapus nomor yang memiliki 6 digit (mengganti dengan value empty) tapi depannya 62

zoho_df.loc[
    (zoho_df['Mobile'].str.len() == 6) & 
    (zoho_df['Mobile'].str[:2] == '62'), 
    'Mobile'
] = 'empty'

zoho_df.loc[
    (zoho_df['Phone'].str.len() == 6) & 
    (zoho_df['Phone'].str[:2] == '62'), 
    'Phone'
] = 'empty'
```

```python
# membersihkan format phone number dan email

m.clean_number(zoho_df, 'Phone')
m.clean_number(zoho_df, 'Mobile')
m.replace_not_valid_to_empty(zoho_df, 'Phone')
m.replace_not_valid_to_empty(zoho_df, 'Mobile')
zoho_df['Email'] = m.lowercase(zoho_df, 'Email')
m.clean_space(zoho_df, 'Email')

count_deleted_wrong_fromat_zoho = m.delete_email_wrong_format(zoho_df, 'Email')
count_after_deleted_wrong_format_zoho = zoho_df.shape[0]

m.clean_number(df_booking, 'Phone1')
m.clean_number(df_booking, 'Mobile')
df_booking['Email'] = m.lowercase(df_booking, 'Email')
m.clean_space(df_booking, 'Email')

# syntax ini harus dimatikan atau diganti deleted email zoho saja agar jumlah row di data booking tidak berkurang (karena ada baris yang terhapus karena format email salah)
count_deleted_wrong_fromat_zoho = m.delete_email_wrong_format(zoho_df, 'Email')
count_after_deleted_wrong_format_zoho = zoho_df.shape[0]

df_booking
```

```python
count_zoho_in_booking_phone = zoho_df.loc[
    (zoho_df['Email'].isin(df_booking['Email'])) &
    (zoho_df['Phone'] != 'empty') & 
    (~zoho_df['Phone'].str.contains('$00000', regex=True)) &
    (zoho_df['Phone'].str.len() < 15) &
    (zoho_df['Phone'].str.len() > 5)
].shape[0]

zoho_in_booking_phone = zoho_df.loc[
    (zoho_df['Email'].isin(df_booking['Email'])) &
    (zoho_df['Phone'] != 'empty') &
    (~zoho_df['Phone'].str.contains('$00000', regex=True)) & 
    (zoho_df['Phone'].str.len() < 15) &
    (zoho_df['Phone'].str.len() > 5)
].copy()

count_zoho_in_booking_mobile = zoho_df.loc[
    (zoho_df['Email'].isin(df_booking['Email'])) &
    (zoho_df['Mobile'] != 'empty') &
    (~zoho_df['Mobile'].str.contains('$00000', regex=True)) & 
    (zoho_df['Mobile'].str.len() < 15) &
    (zoho_df['Mobile'].str.len() > 5)
].shape[0]

zoho_in_booking_mobile = zoho_df.loc[
    (zoho_df['Email'].isin(df_booking['Email'])) &
    (zoho_df['Mobile'] != 'empty') &
    (~zoho_df['Mobile'].str.contains('$00000', regex=True)) & 
    (zoho_df['Mobile'].str.len() < 15) &
    (zoho_df['Mobile'].str.len() > 5)
].copy()
```

```python
zoho_in_booking_phone.replace('empty', '', inplace=True)
zoho_in_booking_mobile.replace('empty', '', inplace=True)
```

```python
list_phone_zoho = zoho_in_booking_phone[['Email', 'Phone']].copy()
list_mobile_zoho = zoho_in_booking_mobile[['Email', 'Mobile']].copy()
```

```python
list_phone_zoho.rename(columns = {'Phone':'Phone Zoho'}, inplace = True)
list_mobile_zoho.rename(columns = {'Mobile':'Mobile Zoho'}, inplace = True)
```

```python
# join table
dfs = [df_booking, list_phone_zoho, list_mobile_zoho]
joined_df = reduce(lambda  left,right:  pd.merge(left,
                     right,
                     left_on ='Email',
                     right_on ='Email',
                     how ='left'), dfs).fillna('empty')


joined_df
```
