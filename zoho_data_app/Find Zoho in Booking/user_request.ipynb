```python
# define data member dan booking yang ingin dijadikan acuan

booking_path = r'C:\Users\fajar\Documents\Python\Data\member-dedup.xlsx'
df_booking = pd.read_excel(booking_path, dtype='string', sheet_name='Sheet1')
df_booking
```

```python
zoho_df.fillna('empty', inplace=True)
zoho_df.reset_index(drop=True, inplace=True)
zoho_df
```

```python
# membersihkan format phone number dan email

m.clean_number(zoho_df, 'Phone')
m.clean_number(zoho_df, 'Mobile')
m.replace_not_valid_to_empty(zoho_df, 'Phone')
m.replace_not_valid_to_empty(zoho_df, 'Mobile')
zoho_df['Email'] = m.lowercase(zoho_df, 'Email')
m.clean_space(zoho_df, 'Email')

count_deleted_wrong_fromat_zoho = m.delete_email_wrong_format(zoho_df, 'Email')
count_after_deleted_wrong_format_zoho = zoho_df.shape[0]

df_booking['Email'] = m.lowercase(df_booking, 'Email')
m.clean_space(df_booking, 'Email')

df_booking
```

```python
zoho_df.groupby('Email').count()
```

```python
df_email = df_booking.loc[df_booking['Same Value Find'] == 'Dedup-Email'].copy()
df_mobile = df_booking.loc[df_booking['Same Value Find'] == 'Dedup-Mobile'].copy()
df_mobile
```

```python
col_email_list = ['Email','Lead_Source_Description','Lead_Locations','Lead_Regions','Lead_Source','Lead_Sub_Brand','Lead_Brand']
email_list = zoho_df.loc[(zoho_df['Lead_Brand'] != 'empty') & (zoho_df['Email'] != 'empty')].copy()
email_list.drop_duplicates(subset=['Email'], inplace=True)
email_list = email_list.reindex(columns=col_email_list)

col_phone_list = ['Phone','Lead_Source_Description','Lead_Locations','Lead_Regions','Lead_Source','Lead_Sub_Brand','Lead_Brand']
phone_list = zoho_df.loc[(zoho_df['Lead_Brand'] != 'empty') & (zoho_df['Phone'] != 'empty')].copy()
phone_list.drop_duplicates(subset=['Phone'], inplace=True)
phone_list = phone_list.reindex(columns=col_phone_list)

col_mobile_list = ['Mobile','Lead_Source_Description','Lead_Locations','Lead_Regions','Lead_Source','Lead_Sub_Brand','Lead_Brand']
mobile_list = zoho_df.loc[(zoho_df['Lead_Brand'] != 'empty') & (zoho_df['Mobile'] != 'empty')].copy()
mobile_list.drop_duplicates(subset=['Mobile'], inplace=True)
mobile_list = mobile_list.reindex(columns=col_mobile_list)
mobile_list
```

```python
# # join table
# dfs = [df_booking, list_phone_zoho, list_mobile_zoho]
# joined_df = reduce(lambda  left,right:  pd.merge(left,
#                      right,
#                      left_on ='Email',
#                      right_on ='Email',
#                      how ='left'), dfs).fillna('empty')


# joined_df

# join table
dfs_email = [df_booking, email_list]
joined_df_email = reduce(lambda  left,right:  pd.merge(left,
                     right,
                     left_on ='Email',
                     right_on ='Email',
                     how ='left'), dfs_email).fillna('empty')

joined_df_email.loc[joined_df_email['Lead_Brand'] != 'empty']
```

```python
dfs_mobile = [df_booking, phone_list]
joined_df_mobile = reduce(lambda  left,right:  pd.merge(left,
                     right,
                     left_on ='Phone',
                     right_on ='Phone',
                     how ='left'), dfs_mobile).fillna('empty')

joined_df_mobile = joined_df_mobile.loc[joined_df_mobile['Lead_Brand'] != 'empty'].copy()
joined_df_mobile
```

```python
dfs_mobile = [df_booking, phone_list]
joined_df_mobile_2 = reduce(lambda  left,right:  pd.merge(left,
                     right,
                     left_on ='Mobile',
                     right_on ='Phone',
                     how ='left'), dfs_mobile).fillna('empty')
joined_df_mobile_2 = joined_df_mobile_2.loc[joined_df_mobile_2['Lead_Brand'] != 'empty'].copy()
joined_df_mobile_2
```

```python
dfs_mobile_2 = [df_booking, mobile_list]
joined_df_mobile_3 = reduce(lambda  left,right:  pd.merge(left,
                     right,
                     left_on ='Phone',
                     right_on ='Mobile',
                     how ='left'), dfs_mobile_2).fillna('empty')

joined_df_mobile_3 = joined_df_mobile_3.loc[joined_df_mobile_3['Lead_Brand'] != 'empty'].copy()
joined_df_mobile_3
```

```python
dfs_mobile_2 = [df_booking, mobile_list]
joined_df_mobile_4 = reduce(lambda  left,right:  pd.merge(left,
                     right,
                     left_on ='Mobile',
                     right_on ='Mobile',
                     how ='left'), dfs_mobile_2).fillna('empty')

joined_df_mobile_4 = joined_df_mobile_4.loc[joined_df_mobile_4['Lead_Brand'] != 'empty'].copy()
joined_df_mobile_4
```

```python
combine = [joined_df_email, joined_df_mobile, joined_df_mobile_2, joined_df_mobile_3, joined_df_mobile_4]
final_df = pd.concat(combine)
final_df.replace('empty', '', inplace=True)
final_df.to_csv(r'C:\Users\fajar\Documents\Python\Data\final_df.csv', index=False)
```
