```python
# CARA INI HANYA EFEKTIF DILAKUKAN JIKA SEMUA KOLOM DI SETIAP SHEET FROMATNYA SUDAH SAMA
# JIKA FORMATNYA KOLOMNYA BERBEDA, LAKUKAN DENGAN CARA SEPERTI DI edm_plan_app

import pandas as pd

file_path = 'EDM Planning & Schedule - CRM (1).xlsx'
```

```python
all_dfs = pd.read_excel(file_path, sheet_name=None)
df = pd.concat(all_dfs, keys=all_dfs,
    names=['Sheets Name', None],
    ).reset_index(level='Sheets Name')   
df.reset_index(drop=True, inplace=True)
df
```

```python
df.columns
```

```python
df.loc[(df['EDM Programme Name'].isna()) & df['Programme Name'].notna(), 'EDM Programme Name'] = df['Programme Name']
df.loc[(df['EDM Programme Name'].isna()) & df['EDM Program Name'].notna(), 'EDM Programme Name'] = df['EDM Program Name']
df.loc[(df['EDM Programme Name'].isna()) & df['EDM Name'].notna(), 'EDM Programme Name'] = df['EDM Name']

df.loc[(df['MailWizz ID for CRM Staff'].isna()) & df['MailWizz ID By CRM Staff'].notna(), 'MailWizz ID for CRM Staff'] = df['MailWizz ID By CRM Staff']
df.loc[(df['MailWizz ID for CRM Staff'].isna()) & df['MailWizz ID by CRM Staff'].notna(), 'MailWizz ID by CRM Staff'] = df['MailWizz ID by CRM Staff']


df.loc[(df['Email Subject Line'].isna()) & df['Subject Line'].notna(), 'Email Subject Line'] = df['Subject Line']
df.loc[(df['PIC'].isna()) & (df['PIC/Approval'].notna()), 'PIC'] = df['PIC/Approval']
```

```python
df = df.reindex(columns=['Lead Brand','Lead Sub Brand','EDM Programme Name', 'Database/Listname', 'Email Subject Line', 'Weekly Digest', 'Date', 'Time', 'Requestor',
       'PIC', 'Approval', 'Status by CRM Staff', 'MailWizz ID for CRM Staff', 'Live Stats for CRM Staff', 'URL Match to HTML', 'EDM Reports','Sheets Name'])
```
