```python
import pandas as pd

# merubah group menjadi dataframe menggunakan reset_index, dan membuat nama kolom baru juga kolom grouping tidak bernama

# creating dataframe
df = pd.DataFrame({'Student': ['Rangga', 'Sucipto', 'Ryan', 'Ahmad', 'Rangga', 'Ryan', 'Sucipto', 'Ahmad', 'Rangga', 'Sucipto', 'Ahmad', 'Ryan'],
                   'Status': [
                   'Siswa Tetap', 'Pertukaran Pelajar', 'Pertukaran Pelajar', 'Siswa Tetap', 'Pertukaran Pelajar', 'Siswa Tetap', 'Pertukaran Pelajar', 'Pertukaran Pelajar', 'Siswa Tetap', 'Pertukaran Pelajar', 'Siswa Tetap', 'Pertukaran Pelajar'],
                    'Amount': [
                   10, 20, 30, 40, 20, 60, 40, 80, 30, 60, 120, 90]})
df
```

```python
# split excel berdasarkan groupby
# assume 'df' is your DataFrame and 'column_name' is the column you want to split on
for name, group in df.groupby('Status'):
    group.to_csv(path+f"{name}.csv", index=False)
```
