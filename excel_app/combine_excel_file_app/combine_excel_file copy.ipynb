```python
import os
import pandas as pd
 
# Get the list of all files and directories
# dir_path = "C://Users//Vanshi//Desktop//gfg"
dir_path = r"C:\Users\fajar.fatoni\Documents\Python\Data\Beambox"

# dir_list = os.listdir(dir_path)
 
# print("Files and directories in '", dir_path, "' :")
 
# # prints all files
# print(dir_list)

frames = []

# itterate file di dalam directory
for path in os.listdir(dir_path):

    # cek apakah di dalam directory ada file
    if os.path.isfile(os.path.join(dir_path, path)):
        if ".xlsx" in path:
            print(path)
            df = pd.read_excel(dir_path+'\\'+path, dtype='string')
            df['filename'] = path
            frames.append(df)

        if ".csv" in path:
            print(path)
            df = pd.read_csv(dir_path+'\\'+path, dtype='string', on_bad_lines='skip')
            df['filename'] = path
            frames.append(df)
    else:
        print('file doesn\'t exist')

result = pd.concat(frames, ignore_index=True)
result
```

```python
result.drop_duplicates(subset=['Email'], inplace=True)
```

```python
result.to_excel(r'C:\Users\fajar.fatoni\Documents\Python\Data\Beambox\df_beambox.xlsx', index=False)
```
