```python
# import OS module
import os
import pandas as pd
 
# Get the list of all files and directories
# dir_path = "C://Users//Vanshi//Desktop//gfg"
dir_path = r"C:\Users\fajar\Documents\Python\Data\Arrival List Raw\Compile Arrival"

count = 0
frames = []
file_name = []
# itterate file di dalam directory
for path in os.listdir(dir_path):
    df_path = dir_path+'\\'+path
    count += 1

    # cek apakah di dalam directory ada file
    if os.path.isfile(os.path.join(dir_path, path)):
        print(df_path)
        globals()['df%s' % count] = pd.read_excel(df_path, sheet_name='Sheet1')
        # memasukan semua file
        frames.append(globals()['df%s' % count])
        file_name.append(path)
    else:
        print('file doesn\'t exist')
    
# reset index dan mengkonversi key grouping menjadi kolom
df = pd.concat(
    frames, 
    keys=file_name,
    names=['Filename', None], # None dapat dihilangkan atau tidak digunakan (Karena none hanya berfungsi untuk menamai kolom index)
).reset_index(level='Filename') # Melakukan reset index urutan paling awal (dapat juga menggunakan value nomor index seperti 0)

df.reset_index(drop=True, inplace=True)

df.columns
```

```python
df = df.reindex(columns=['Sheet Row No.', 'Programme', 'Nationality', 'Region',
       'Client Name', 'Email', 'Phone', 'Check in', 'Check out', 'Sheet Name', 'Filename'])
df
```
